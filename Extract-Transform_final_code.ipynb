```python
# Get the backers_info from the crowdfunding_info sheet. 
pd.set_option('max_colwidth', 400)
backers_info = pd.read_csv('backer_info.csv')

display(backers_info)
```

```python
## In the directions, and what was given on the module information, it was requested to convert the dataframe into a dictionary
## and THEN iterate through, creating a LIST of VALUES. I read this as an actual list, containing the VALUES (ie:key:value pairing)
## and created this from there. I went ahead and cleaned up the lists in the first function.


# Iterate through the backers DataFrame and convert each row to a dictionary.
backers = backers_info['backer_info'].to_dict()

# An alternate method of this is this:
# backers_value=[]
#     # Iterate through each dictionary (row) and get the values for each row using list comprehension.
# for i, row in backer_df.iterrows():
#     data=row['backer_info']
#     converted_data=json.loads(data)

backer_id = [backer_id for backer_id in backers.values()]
backer_id

backer_id_list = []
cf_id_list = []
name_list = []
email_list = []
# """
# -function for parsing the converted dataframe into dictionaries values. Create an initial sublist to remove :, leveraging 
# this on the next line by splitting the key value pairs by the newly established ','. For loop parses the length of the 
# list. For each established list of values we are trying to obtain, we are searching for their unique values using indexes, and 
# then cleaning the data based on the dataframe above.
# """
def butts(): 
    for backer in backers.values():
        sub_list = backer.replace(":", ",")
#         print(sub_list)
        sl2 = sub_list.split(',')
#         print(sub_list_2)
        for index in range(1, len(sl2), 2):
#             print(sl2[1])
            backer_id_list.append(sl2[1].replace('"','').replace(' ',''))
            cf_id_list.append(sl2[3].replace(' ',''))
            name_list.append(sl2[5].replace('"','').lstrip())
            email_list.append(sl2[7].replace('"','').replace('}','').replace(' ',''))
    
butts()
dictionary_list = []
# print(backer_id_list)
def beepbeep(backer_id_list, cf_id_list, name_list, email_list):
# """
# -For statement takes in four different list of values found and extracts the values and put them into a new list of dictionaries
# with their values and keys automatically paired up 
# -x=0 statement and if x>10, acts as a control statement so when debugging our code we don't have to iterate through
# thousands of lines in the dataframe. The code exits after ten iterations, and returns the results so that we can configure the 
# rest of the logic to clean up the code.
# """
#     x=0
    for index in range(len(backer_id_list)):
        new_dictionary = {'backer_id': backer_id_list[index], 'cf_id': cf_id_list[index], 'name': name_list[index], 'email': email_list[index]}
        dictionary_list.append(new_dictionary)
#         if x>10:
#             break
#         else:
#             x+=1
#         print(new_dictionary)
beepbeep(backer_id_list, cf_id_list, name_list, email_list)


## Mitigating/ cleaning up the action of the for loops/ensuring that there are no duplicates across all the rows.
## Set() unlike lists or tupes cannot have multiples occurrences of the same element. Key established as iterator iterating 
## 'backer_id' and 'cf_id'. We want two columns because there may be certain values that repeat on purpose, so you want to
## choose values that are purposely unique like the name, or the id. These are used to control any duplicates because you 
## cannot store the same value twice in a set. Once the ID or name is found again, it skips over that record, effectively
## deleting the duplicate. Then we add any unique records to the results list, and thats the new duplicate free list of values.

seen = set()
result = []
for dic in dictionary_list:
    key = (dic['backer_id'], dic['cf_id'])
    if key in seen:
        continue
    result.append(dic)
    seen.add(key)
    

# print(result)
# print(dictionary_list[:10])
#Check that its an actual dictionary and not a string of shit :) 
# print(type(dictionary_list[0]))

#Iterate through each dictionary (row) and get the values for each row using list comprehension.
# Append the list of values for each row to a list. 
values_list = []
for x in result:
    values=x.values()
    itdr = [x for x in values]
    values_list.append(itdr)
print(values_list[:100])


# Append the list of values for each row to a list.
# An alternative method of this is:
#     backer_values.append(converted_data)
# # Print out the list of values for each row.
# print(backer_value)
    
##LSTRIP IS PRETTY IMBALANCED##





```

```python
# Create a backers_df DataFrame with the following columns: 'backer_id','cf_id', 'name', and 'email' 
# using the list of values lists. df = pd.DataFrame (list_name, columns = ['column_name'])
backers_df = pd.DataFrame(values_list, columns = ['backer_id','cf_id','name','email'])

# backers_df.drop(columns='index')
backers_df.head(10)

```
